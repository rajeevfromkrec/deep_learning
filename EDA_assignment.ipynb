#loading data



```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```
data=pd.read_csv('haberman.csv')
```

```
data.head(20)
```

#Plotting

```
sns.set(style="whitegrid")
sns.scatterplot(x='age',y='nodes', data=data)
```

1.There does not seem to be much co-relation beyween nodes and Age
2.Majority of patients are between age 40 and 60


```
sns.set(style="whitegrid")
sns.scatterplot(x='age',y='nodes',hue='status', data=data)
```

```
data['status'].value_counts()
```



1.  Number of patients who died in less than 5 years(2) are less than number of patients died after 5 years(1)
2.  Number of patients who lived more than 5 years had less nodes.
3. There seems to be some corelation between number of nodes and class label 3



```
data['status'].isnull().sum()
```

```
sns.set(style="whitegrid")
sns.scatterplot(x='year',y='age',hue='status', data=data)
```

```

g = sns.FacetGrid(data, col="nodes", height=4, col_wrap=5)
g.map(plt.hist, "status")
```



*   Maximun number of patients are with nodes 0,1,2,3
